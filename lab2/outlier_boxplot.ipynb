```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaled_array = scaler.fit_transform(tips[['total_bill', 'tip']])
scaled_array
```

```python
import pandas as pd
scaled_df = pd.DataFrame(
    scaled_array,
    columns=['total_bill', 'tip'],
    index=tips.index
)
```

```python
scaled_df
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
tips
```

```python
one_hot = OneHotEncoder()
one_hot_sex = one_hot.fit_transform(tips[['sex']])
```

```python
one_hot_sex.toarray()
```

```python
one_hot_sex_df = pd.DataFrame(
    one_hot_sex.toarray(),
    columns=one_hot.get_feature_names_out(),
    index=tips.index
)
```

```python
one_hot_sex_df
```

```python
final_df = pd.concat([tips[['sex']], one_hot_sex_df], axis=1)
final_df
```

```python
label_encoder = LabelEncoder()
label_sex = label_encoder.fit_transform(tips['sex'])
```

```python
label_sex
```

```python
label_sex_df = pd.DataFrame(
    label_sex,
    columns=['sex'],
    index=tips.index
)

final_label_sex_df = pd.concat([tips[['sex']], label_sex_df], axis=1)
```

```python
final_label_sex_df
```
